```python
data['Ship Mode'].value_counts()
```

```python
x = data['Ship Mode'].value_counts().index
y = data['Ship Mode'].value_counts().values
y
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%",shadow=2,)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Ship Mode",data=data,hue="Category")
plt.title("Shipping mode by product category ")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Segment",data=data)
plt.title("figure by customer segment")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Category",data=data)
plt.title("category")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Category",data=data[data["Category"]=="Furniture"],hue="Sub-Category")
plt.title("category furniture based sub_category")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="Category",data=data[data["Category"]=="Office Supplies"],hue="Sub-Category")
plt.title("category Office Supplies based sub_category")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Category",data=data[data["Category"]=="Technology"],hue="Sub-Category")
plt.title("category Office Supplies based sub_category")
plt.show()
```

```python
data["Order Year"] = data["Order Date"].dt.year
```

```python
data["Order Year"].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Year",data=data,palette="pastel")
plt.title("Increase by order on yearly")
plt.show()
```
